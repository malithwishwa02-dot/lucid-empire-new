```python
# Import and path setup
import sys
import os
import importlib.util

ROOT = r"e:/final/camoufox-main"
if ROOT not in sys.path:
    sys.path.append(ROOT)

modules_to_check = [
    ("core.genesis_engine", "Genesis Engine (Heart)"),
    ("modules.commerce_injector", "Commerce Module (Nerves)"),
    ("modules.biometric_mimicry", "Biometric Module (Motor Control)"),
    ("dashboard.main", "Dashboard (Brain)")
]

results = []

for module_path, description in modules_to_check:
    try:
        spec = importlib.util.find_spec(module_path)
        if spec is None:
            raise ImportError(f"spec not found for {module_path}")
        results.append((module_path, description, True, None))
        print(f"  [PASS] {description:30} ... SPEC FOUND")
    except Exception as e:
        results.append((module_path, description, False, str(e)))
        print(f"  [FAIL] {description:30} ... ERROR: {e}")

fails = [r for r in results if not r[2]]
print("\nSUMMARY:")
print(f"  Total checks: {len(results)}; Failures: {len(fails)}")
if fails:
    print("\nFailures detail:")
    for f in fails:
        print(f" - {f[1]} ({f[0]}): {f[3]}")
    raise SystemExit(2)
else:
    print("[OK] All module specs detected.")
```
